```python
from keras.models import Sequential
from keras.layers import Dense

import numpy as np
import tensorflow as tf
```

```python
pd.read_csv('ThoraricSurgery.csv',header=-1)
```

```python
# seed 를 통해 0보다  크거나 같은 임의의 정수를 추출(동일 seed 는 같은 난수가 발생 )
seed =0
np.random.seed(seed)

tf.set_random_seed(seed)
Data_set= np.loadtxt('ThoraricSurgery.csv',delimiter=',')
```

```python
#전체 데이터 행 중에서 0번째부터 17번째 열 갓ㅂ들을 x리스트에 저장
X=Data_set[:,0:17]
#전체 데이터에서 17번째 여값을 y리스트에 저장
y=Data_set[:,17]
```

```python
X
```

```python
#신경망을 한층 한층 쌓는 기능 (add()함수 사용 )
model=Sequential()
#Dense():신경망 각층의 특성을 설정하는 기능 
model.add(Dense(30,input_dim=17,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# compile 학습 프로세스를 설정하는 기능 
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
model.fit(X,y,epochs=30,batch_size=10)
```

```python
#결과 출력

print ('\n 정확도 :',model.evaluate(X,y)[1])
```
