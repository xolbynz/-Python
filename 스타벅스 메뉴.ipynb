```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

```python
url="http://www.istarbucks.co.kr/menu/drink_list.do"
driver=wb.Chrome()
driver.get(url)
```

```python
a2=driver.find_element_by_css_selector('#container > div.content > div.product_result_wrap.product_result_wrap01 > div > dl > dt.dt2 > a')
```

```python
a2.click()
```

```python
soup=bs(driver.page_source,'lxml')
```

```python
soup
```

```python
drink1=soup.select('td')
```

```python
drink1
```

```python
drink_text=[]
for drink in range(len(drink1)):
    drink_text.append(drink1[drink])
```

```python
drink_text
```

```python
len(drink1)
```

```python
drink_menu=[]
drink_Calorie=[]
drink_sugars=[]
drink_protein=[]
drink_salt=[]
drink_Saturated_fat=[]
drink_Caffeine=[]
for drink in range(0,len(drink1),7):
    drink_menu.append(drink1[drink].text)
    drink_Calorie.append(drink1[drink+1].text)
    drink_sugars.append(drink1[drink+2].text)
    drink_protein.append(drink1[drink+3].text)
    drink_salt.append(drink1[drink+4].text)
    drink_Saturated_fat.append(drink1[drink+5].text)
    drink_Caffeine.append(drink1[drink+6].text)
```

```python
dic={'메뉴':drink_menu,'칼로리(kal)':drink_Calorie,'당류(g)':drink_sugars,'단백질(g)':drink_protein,'나트륨(g)':drink_salt,'포화지방(g)':drink_Saturated_fat,'카페인(mg)':drink_Caffeine}
```

```python
starbuck_menu=pd.DataFrame(dic)
```

```python
starbuck_menu=starbuck_menu[['메뉴','칼로리(kal)','당류(g)','단백질(g)','나트륨(g)','포화지방(g)','카페인(mg)']]
```
