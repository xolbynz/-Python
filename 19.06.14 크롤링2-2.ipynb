```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup as bs
```

# 날짜 생성

```python
date = pd.date_range(end = '2019-06-14', periods = 100)
date[0].strftime('%Y%m%d')
```

```python
day_list = []
for i in date:
    day_list.append(i.strftime('%Y%m%d'))
```

```python
day_list
```

```python
Mday_list=[]
title_list = []
rate_list = []
for date in day_list:
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='+date
    res = req.get(url)
    soup = bs(res.content,'lxml') 
    Mrank = soup.find_all("div", class_='tit5')
    Mrank2 = soup.find_all("td", class_='point')
    for title in Mrank:
        title_list.append(title.text.strip())
        Mday_list.append(date)
    for rate in Mrank2:
        rate_list.append(rate.text.strip())
    
```

```python
dic = {'date' : Mday_list ,'title' : title_list, 'rate':rate_list}
Movie = pd.DataFrame(dic)
Movie
```

```python
Movie.info()
```

```python
Movie['date']=Movie['date'].apply(pd.to_datetime)
Movie['rate']=Movie['rate'].astype('float')
```
