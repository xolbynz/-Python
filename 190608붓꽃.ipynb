```python
import numpy as py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(iris_df)
```

```python
kmeans.labels_
```

```python
kmeans.predict(iris_df)
```

```python
iris_df["label"]=kmeans.labels_
```

```python
sns.pairplot(iris_df, hue="label", vars=iris.feature_names)
```

# 스케일링

```python
from sklearn.preprocessing import StandardScaler
```

```python
iris_df2= pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
sc=StandardScaler()
sc.fit_transform
```

```python
df=sc.fit_transform(iris_df2)
```

```python
iris_df3= pd.DataFrame(df, columns=iris.feature_names)
```

```python
iris_df3
```

```python
kmeans= KMeans(n_clusters=3)
kmeans.fit(iris_df3)
```

```python
iris_df3["label"]=kmeans.predict(iris_df3)
```

```python
sns.pairplot(iris_df3, hue="label",vars=iris.feature_names)
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
adjusted_rand_score(iris.target,iris_df["label"])
```

```python
adjusted_rand_score(iris.target,iris_df3["label"])
```

# pipline 사용

```python
iris_df4= pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
sc=StandardScaler()
kmeans= KMeans(n_clusters=3)
```

```python
pipline=make_pipeline(sc, kmeans)
```

```python
pipline.fit(iris_df4)
```

```python
pipline.predict(iris_df4)
```
