```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
```

```python
import time
```

```python
driver = wb.Chrome()
url='https://www.istarbucks.co.kr/store/store_map.do?disp=locale'

driver.get(url)
time.sleep(2)
```

```python
btn_search1=driver.find_element_by_class_name('set_sido_cd_btn')
btn_search1.click()
time.sleep(2)
```

```python
btn_search2=driver.find_element_by_css_selector('ul.gugun_arae_box li+li+li+li+li+li+li+li+li+li+li+li+li+li a.set_gugun_cd_btn')
btn_search2.click()
time.sleep(2)
```

```python
soup=bs(driver.page_source,'lxml')
```

```python
soup
```

```python
location=soup.select('p.result_details')
```

```python
subway=soup.select('ul.quickSearchResultBoxSidoGugun strong')
```

```python
location_list=[]
subway_list=[]
tel_list=[]
```

### 데이터 전화번호와 정소를 구분지으려고하는데, 전화번호 길이가 다 달라서 02 앞으로 주소 로 지정하고 02뒤로 전화번호로 지졍  

```python
for i in range(len(location)):
    
    subway_list.append(subway[i].text)
    location_list.append(location[i].text.split('02-')[0])
```

```python
location_list
```

02뒤로ㅓ 짜르고 02 다시붙친다.

```python
for i in range(len(location)):
    tel_list.append('02-'+(location[i].text.split('02-')[1]))  
```

```python
tel_list
```

```python
dic={"매장주소":location_list,"매장명":subway_list,"전화번호":tel_list}
```

```python
import pandas as pd
```

```python
starbuck=pd.DataFrame(dic)
```
