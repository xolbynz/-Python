```python
import time
```

```python
try:  # 자바에서 트라이케치문과 같다. 트라이를 시도하고 오류가 뜨면 익셉션으로 내려간다. 
    for ss in range(50):
        input_search=driver.find_element_by_class_name('c_05') # 더보기 를 50번 클릭하여 끝까지 메뉴를 본다  끝이나면 익셉션으로 내려간다.
        input_search.click()
        time.sleep(3) # 포문은 빠르고 더보기반응은 느려서 포문을 느리게돌리게한다.
except:
    print(ss)#페이지수를 확인하고 끝을낸다.
```

```python
soup=bs(driver.page_source,'lxml')
```

```python
menu_list=[]
price_list=[]
menu=soup.select('div.item-text h4')
price=soup.select('div.item-text strong')
for i in menu:
    menu_list.append(i.text.split())
for j in price:
    price_list.append(j.text.split())
```

```python
menu_list
```

#### 띄어쓰기 반찬 등 때문에 다차원 배열이 되어ㅇ;ㅆ다.
조인을 이용한 합치기

```python
realmenu_list=[]
for i in menu_list:
    realmenu_list.append(" ".join(i))
```

```python
import pandas as pd
```

```python
realprice_list=[]
for i in price_list:
    realprice_list.append(" ".join(i))
```

#### 딕셔너리를 판다스 데이터 프레임에 넣으려고한다. 딕셔너리 열 을 만들고 데이터(가격정보와메뉴정보)를 넣는다 

```python
dic1={"메뉴":realmenu_list,"가격":realprice_list}
```

```python
dic1
```

```python
hansot=pd.DataFrame(dic1)
```

```python
hansot=hansot[["메뉴","가격"]] 
```

```python
hansot
```

```python
hansot.to_csv('hansot.csv', encoding='EUC-KR' , index=False)
```
