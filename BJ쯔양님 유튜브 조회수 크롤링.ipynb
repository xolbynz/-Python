```python
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.youtube.com/channel/UCfpaSruWW3S4dibonKXENjA'
driver.get(url)
```

```python
driver.find_element_by_css_selector('#tabsContent > paper-tab:nth-child(4)').click()
```

스크롤을 내려서 동영상을 다 보자 !

try except 문을 사용하는 방법을 알아보고싶은데 모르겠다 그냥 스크롤을 200번 내려서 찾아보자 

```python
body=driver.find_element_by_tag_name('body')
for i in range(200):
    body.send_keys(Keys.PAGE_DOWN)
   
```

```python
soup=bs(driver.page_source,'lxml')
```

```python
movie=soup.select('a#video-title')
```

영상 타이틀 추출

```python
movie_list=[]
for m in movie:
    movie_list.append(m.text)
```

```python
movie_list
```

영상 url 추출 

```python
movie_img=soup.select('#video-title')
```

```python
movie_img[30].get('href')
#get을사용하여 href 태그를 찾아올수 있다.
```

```python
movie_img_list=[]
for c in range(0,len(movie_img)):
    movie_img_list.append('http://www.youtube.com'+movie_img[c].get('href'))
```

```python
movie_img_list
```

```python
movie_join_date=soup.select('#metadata-line > span')
```

```python
movie_join_list=[]
for index in range(0,len(movie_join_date),2):
    movie_join_list.append(movie_join_date[index].text[4:])
```

동영상 갯수 획인 .한줄이면 30개이고 전체는  149개

```python
print(len(movie_list))
print(len(movie_join_list))
print(len(movie_img_list))
```
